Для анализа нам понадобится рассчитать расстояние Левенштейна

```python
!pip install python-Levenshtein
from Levenshtein import distance as lev
import pandas as pd
import re
```

Подгрузим словарь

```python
result = pd.read_csv('словарь.csv')
result = result[['слово', 'перевод', 'источник', 'словарь', 'слово латиница']]
result
```

Протестируем на фамилиях

```python
# пишем в name любую фамилию и проверяем
name = 'Дерменджи'

subset = result.copy()
subset['Lev'] = subset['слово'].apply(lambda x: lev(name.lower(), x.lower()))
subset = subset.sort_values(['Lev'])
subset
```

Кириллизация

```python
def custom_cyrillicize(token):
        cyrillic_mapping = {'А':'А', 'а':'а', 'Ä':'Я', 'ä':'я', 'B':'Б', 'b':'б', 'C':'Дж', 'c':'дж',
                            'Ç':'Ч', 'ç':'ч', 'D':'Д', 'd':'д', 'Е':'Е', 'е':'е', 'Ê':'Э', 'ê':'э',
                            'F':'Ф', 'f':'ф', 'G':'Г', 'g':'г', 'H':'Х', 'h':'х', 'I':'И', 'ı':'и',
                            'İ':'И', 'i':'и', 'J':'Ж', 'j':'ж', 'K':'К', 'k':'к', 'L':'Л', 'l':'л',
                            'M':'М', 'm':'м', 'N':'Н', 'n':'н', 'O':'О', 'o':'о', 'Ö':'ё', 'ö':'ё',
                            'Р':'П', 'р':'п', 'R':'Р', 'r':'р', 'S':'С', 's':'с', 'Ş':'Ш', 'ş':'ш',
                            'T':'Т', 't':'т', 'Ţ':'Ц', 'ţ':'ц', 'U':'У', 'u':'у', 'Ü':'ю', 'ü':'ю',
                            'V':'В', 'v':'в', 'Y':'Й', 'y':'й', 'Z':'З', 'z':'з'}

        cyrillicized_token = ''.join(cyrillic_mapping.get(char, char) for char in token)
        return cyrillicized_token
```

Выводим значение

```python
df = pd.DataFrame(subset.head(1))

# Достаем значение перевода
перевод_значение = df['перевод'].values[0]
словарь = df['словарь'].values[0]
словарь1 = re.sub(r'\w{2}$', 'ого', словарь)
# Выводим значение перевода
print("Скорее всего, эта фамилия произошла от", словарь1, "слова", '"',перевод_значение,'"')
```
